```python
import pandas as pd

df = pd.read_excel('./source/data_series.xlsx')
df
```

```python
df = df.dropna()
df
```

```python
df.columns
```

```python
cols = df.columns.tolist()
new_cols = []

i = 0
while i < len(cols):
    col = cols[i]
    if isinstance(col, int): 
        year = col
        new_cols.append(f"{year}_target")
        new_cols.append(f"{year}_realisasi")
        new_cols.append(f"{year}_capaian%")
        i += 3  
    else:
        new_cols.append(col)
        i += 1

df.columns = new_cols

df
```

```python
df.columns
```

```python
df_num = df.copy()

cols_to_convert = df_num.columns.drop('Nama')

df_num[cols_to_convert] = (
    df_num[cols_to_convert]
    .replace({r'\.': '', r',': '.'}, regex=True)
    .apply(pd.to_numeric, errors='coerce')
)

df_num
```

```python
df_num.info()
```

```python
target_cols = ['Nama'] + [c for c in df_num.columns if c.endswith('_target')]
target = df_num[target_cols].rename(
    columns={c: c.replace('_target', '') for c in target_cols if c != 'Nama'}
)

realisasi_cols = ['Nama'] + [c for c in df_num.columns if c.endswith('_realisasi')]
realisasi = df_num[realisasi_cols].rename(
    columns={c: c.replace('_realisasi', '') for c in realisasi_cols if c != 'Nama'}
)

capaian_cols = ['Nama'] + [c for c in df_num.columns if c.endswith('_capaian%')]
persentase_capaian = df_num[capaian_cols].rename(
    columns={c: c.replace('_capaian%', '') for c in capaian_cols if c != 'Nama'}
)
```
